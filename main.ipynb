```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_sum = pd.read_excel(io='stadat-iga0008-11.1.2.1-en.xlsx', sheet_name='total')
df_larceny = pd.read_excel(io='stadat-iga0008-11.1.2.1-en.xlsx', sheet_name='larceny')
df_larceny['type'] = 'larceny'
df_public_nuisance = pd.read_excel(io='stadat-iga0008-11.1.2.1-en.xlsx', sheet_name='public nuisance')
df_public_nuisance['type'] = 'public nuissance'
df_official_person = pd.read_excel(io='stadat-iga0008-11.1.2.1-en.xlsx', sheet_name='official person')
df_official_person['type'] = 'attack against official person'
df_driving_under_inf = pd.read_excel(io='stadat-iga0008-11.1.2.1-en.xlsx', sheet_name='drink drive')
df_driving_under_inf['type'] = 'driving under drug or alcoholic influence'
df_drug = pd.read_excel(io='stadat-iga0008-11.1.2.1-en.xlsx', sheet_name='drug')
df_drug['type'] = 'drug related'
df_battery = pd.read_excel(io='stadat-iga0008-11.1.2.1-en.xlsx', sheet_name='battery')
df_battery['type'] = 'battery'
dataframes = [df_larceny, df_public_nuisance, df_official_person, df_driving_under_inf, df_drug, df_battery]
df_total = pd.DataFrame()
for dataframe in dataframes:
    df_total = pd.concat([df_total, dataframe])
years = [i for i in range(2009,2024)]
years_rates = [str(i) + str('_rate') for i in years]
```

```python
df_popularity = pd.read_excel('stadat-nep0034-22.1.2.1-hu.xlsx')
```

```python
for df in [df_sum, df_total]:
    df.rename(columns={'Name of the territorial unit': 'Territorial unit'}, inplace=True)
```

```python
df_sum[2009]
```

```python
df_total
```

```python
for year in years:
    name = str(year) + str('_rate')
    df_sum[name] = df_sum[year] / df_popularity[year]
```

```python
df_sum
```

```python
df_sum_rates = df_sum[['Territorial unit',*years_rates]].copy()
df_sum_rates = df_sum_rates.melt(id_vars='Territorial unit', var_name='year', value_name='rate')
df_sum_rates
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='Territorial unit', y='rate', data=df_sum_rates)
plt.xticks(rotation=90)
plt.title(f'Crimes per person by territorial unit ({years[0]} - {years[-1]})')
plt.ylabel('Crime per person')
plt.xlabel('Territorial unit')
plt.show()
```

```python
df_sum_years = df_sum[['Territorial unit',*years]].copy()
df_sum_years = df_sum_years.melt(id_vars='Territorial unit', var_name='year', value_name='crimes_count')
df_sum_years
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='year', y='crimes_count', data=df_sum_years)
plt.xticks(rotation=90)
plt.title(f'Crimes by year ({years[0]} - {years[-1]})')
plt.ylabel('Crimes')
plt.xlabel('Year')
plt.show()
```
